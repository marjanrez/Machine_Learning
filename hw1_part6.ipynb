```python
# A.

from sklearn.datasets import make_blobs
from matplotlib import pyplot
from pandas import DataFrame
# generate 2d classification dataset
X, y = make_blobs(n_samples=250, centers=[(-1.0, -1.0),(1,1)], n_features=2, cluster_std=1.0/8, random_state=1)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'r', 1:'b'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key], alpha=0.50, s=10)
pyplot.show()
```

```python
#B.

import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt


mu_x_A = -1.0
mu_x_B = 1.0
mu_y_A = -1.0
mu_y_B = 1.0
sigma = 1.0/8
num_sample = 250

np.random.seed(1234)
x_A = [sigma* np.random.randn() + mu_x_A for i in range(num_sample)]
y_A = [sigma* np.random.randn() + mu_y_A  for i in range(num_sample)]
x_B = [sigma* np.random.randn() + mu_x_B for i in range(num_sample)]
y_B = [sigma* np.random.randn() + mu_y_B for i in range(num_sample)]


plt.plot(x_A, y_A, 'o', color='red',label='classA')
plt.plot(x_B, y_B, 'o', color='blue',label='classB')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
